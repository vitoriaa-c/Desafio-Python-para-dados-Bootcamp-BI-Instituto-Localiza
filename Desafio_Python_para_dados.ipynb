Questão 1. Ao visualizar a base você percebeu que seria melhor alterar o nome de algumas colunas. Mude o ‘ID’ para ‘Identificador’, corrija o nome da coluna que indica a pressão sanguínea, mude a coluna ‘Ocupação’ para ‘Profissão’,  coluna ‘Categoria BMI’ está em parte em inglês, substitua para ‘Categoria IMC’.

```python
# nomes das colunas antes de renomear
dados.columns
```

```python
# rename
dados = dados.rename(columns={'ID': 'Identificador', 'Pressão sanguíneaaaa': 'Pressão sanguínea', 'Ocupação': 'Profissão','Categoria BMI': 'Categoria IMC'})
```

```python
# nomes das colunas após renomeação
dados.columns
```

Questão 2. Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões?

```python
# estatísticas de horas de sono por profissão
dados.groupby('Profissão')['Duração do sono'].agg(['mean', np.median, pd.Series.mode])
```

Questão 3. Das pessoas que atuam com engenharia de software qual a porcentagem de obesos?


```python
# filtro para engenheiros
engenheiros = dados[dados['Profissão'] == 'Eng. de Software']
total = len(engenheiros)

# obesos
obesos = (engenheiros['Categoria IMC'] == 'Obesidade').sum()

# porcentagem
porcentagem = (obesos / total) * 100
print(f"{porcentagem:.2f}%")
```

Questão 4. De acordo com os dados, advogar ou ser representante de vendas faz você dormir menos? (Use o método ‘isin’, considere a média)

```python
# comparando essas profissões, ser representante de vendas faz dormir menos do que advogar, mas comparando com as demais acima, advogar não faz dormir menos
dados[dados['Profissão'].isin(['Representante de Vendas', 'Advogado(a)'])].groupby('Profissão')['Duração do sono'].mean()
```

Questão 5. Entre quem fez enfermagem e quem fez medicina, quem tem menos horas de sono? (Use o método ‘isin’, considere a média)

```python
# quem fez medicina tem menos horas de sono
dados[dados['Profissão'].isin(['Enfermeiro(a)', 'Médico(a)'])].groupby('Profissão')['Duração do sono'].mean()
```

Questão 8.  Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a média)

```python
# separação dos dados da pressão pela barra de acordo com o tipo (veio como texto - object)
dados[['sistólica', 'diastólica']] = dados['Pressão sanguínea'].str.split('/', expand=True)

# conversão de texto para numérico
dados['sistólica'] = pd.to_numeric(dados['sistólica'])
dados['diastólica'] = pd.to_numeric(dados['diastólica'])

# mulheres tem maior pressão sanguínea média
dados.groupby('Gênero')[['sistólica', 'diastólica']].mean()
```

Questão 9. É predominante entre os participantes dormir 8 horas por dia (considere usar Moda como medida)?

Questão 10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
# grupos por frequência
grupo1 = dados[dados['Frequência cardíaca'] > 70]
grupo2 = dados[dados['Frequência cardíaca'] <= 70]
```

```python
# pessoas com frequência cardíaca acima de 70 dão, em média, menos passos que pessoas com frequência cardíaca menor ou igual a 70
print(round(grupo1['Passos diários'].mean(), 2))
print(round(grupo2['Passos diários'].mean(), 2))
```
